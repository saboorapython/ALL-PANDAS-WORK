```python
df["HS"]=df["HS"].str.split(pat="*").str[0]
```

```python
df
```

```python
df.rename(columns={'Span':"career_duration"},inplace=True)
```

```python
df   
```

```python
df["start_year"]=df["career_duration"].str.split(pat="-").str[0]
df["end_year"]=df["career_duration"].str.split(pat="-").str[1]
```

```python
df
```

```python
df.info()
```

```python
df["start_year"]=df["start_year"].astype(int)
df["end_year"]=df["end_year"].astype(int)
```

```python
df
```

```python
df["career_duration"]=df["end_year"]-df["start_year"]
```

```python
df
```

```python
df.groupby("Player")["career_duration"].sum().sort_values(ascending=False).head(5)
```

```python
df.rename(columns={'6s':"sixes"},inplace=True)
```
